```python
# date 열의 데이터 타입을 datetime64로 변경하세요.
df['date'] = pd.to_datetime(df['date'], format='%m-%d-%Y')
df
```

```python
# df의 열 별 데이터 타입을 조회하세요.
df.dtypes
```

```python
# date 열의 년월일 값을 추출하여 새 열 year, month, day를 생성하세요.
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df
```

```python
# year, month, day 열을 이용하여 datetime64 타입의 열 date2를 생성하세요.
df['date2'] = pd.to_datetime(df[['year', 'month', 'day']])
df
```

```python
# df의 열 별 데이터 타입을 조회하세요.
df.dtypes
```

```python
# date2 열을 삭제하세요.
df =df.drop('date2', axis=1)
df
```

```python
# 5월 3일의 데이터만 추출하세요.
df[df['date'].dt.day == 3]
```

```python
# date 열을 인덱스로 지정하세요.
df = df.set_index('date')
df
```

```python
# 5월 3일의 데이터만 추출하세요.
df[df.index.day == 3]
```

```python
# index를 다시 정수로 초기화하세요.
df = df.reset_index()
df
```

```python
# total_bill 값이 높은 순서대로 데이터를 정렬하세요.
df.sort_values('total_bill', ascending=False)
```

```python
# size (방문 인원) 수가 적은 순서대로 데이터를 정렬하세요.
df.sort_values('size')
```

```python
# 성별과 tip 순으로 데이터를 정렬하되, 성별은 오름차순, tip은 내림차순으로 정렬하세요.
df.sort_values(['sex','tip'], ascending=[True, False])
```

```python
# tip을 가장 많이 지불한 고객 상위 10명의 데이터를 출력하세요.
df.sort_values('tip', ascending=False).head(10)
```

```python
# total_bill과 tip의 합계 금액이 가장 낮은 고객 10명의 데이터를 출력하세요.
df.loc[(df['total_bill'] + df['tip']).sort_values().index].head(10)
```

```python
# 흡연자 중 tip을 가장 많이 지불한 상위 10명의 데이터를 추출하세요.
df[df['smoker']=='Yes'].sort_values('tip', ascending=False)
```

```python
# total_bill이 높은 상위 20명의 남녀 수를 출력하세요.
df.sort_values('total_bill', ascending=False).head(20)['sex'].value_counts()
```

```python
# tip이 낮은 하위 20명의 방문 날짜 비율을 출력하세요.
# 날짜 순으로 정렬하세요.
df.sort_values('tip').head(20)['day'].value_counts(normalize=True).sort_index()
```

```python
# total_bill 대비 tip의 비율이 높은 상위 10명의 흡연자 비율을 출력하세요.
df.loc[(df['tip']/df['total_bill']).sort_values(ascending=False).index].head(10)['smoker'].value_counts(normalize=True)
```

```python
# size는 방문 고객의 그룹 인원입니다.
# total_bill은 해당 방문 고객들이 지출한 식사 비용입니다.
# 1인 당 지출한 평균 금액이 적은 순서대로 데이터를 정렬하세요.
df.loc[(df['total_bill']/df['size']).sort_values().index]
```

```python
# 5월 1일에 방문한 고객 중 여성 고객의 데이터를 tip 금액을 기준으로 내림차순 정렬하세요.
cond1 = df['date'].dt.day==1
cond2 = df['sex'] == 'Female'

df[cond1&cond2].sort_values('tip', ascending=False)
```

```python
df[cond1&cond2].sort_values(
    ['smoker', 'size', 'total_bill'], 
    ascending=[False, True, False]).reset_index()
```

```python
# 5월 6일에 방문한 고객 중 남성 고객의 데이터를 
# 흡연자, size, total_bill을 기준으로 각각 내림, 오름, 내림차순으로 정렬하세요.
# 정렬한 결과의 인덱스를 초기화한 후 df2에 저장한 후 출력하세요.
cond1 = df['date'].dt.day==6
cond2 = df['sex'] == 'Male'

df2 = df[cond1&cond2].sort_values(
    ['smoker', 'size', 'total_bill'], 
    ascending=[False, True, False]).reset_index(drop=True)
df2
```
