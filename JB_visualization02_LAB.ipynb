#### 이상치
 * Univariate vs. Multivariate
   - Uni: Boxplot, IQR, 2/3sigma
   - Multi: Scatter, distances
 * Natural
   - 정상적인 outlier
 * Non-natural
   - 기입 오류: 데이터 수집 과정에서 기입을 잘못하는 경우 1000->10000
   - 측정 오류: 측정 장비의 오류로 인한 경우 (일부)
   - 실험 오류: 변인 통제가 제대로 되지 않은 경우
   - 의도적인 오류: 당신의 몸무게는..? 당신의 키는..?
   - 데이터 처리 오류: FE 작업 과정으로 인한 경우
   - 표본 추출 오류: 전국민 수면시간을 계산해야 하는데 영아의 수면 시간 데이터를 샘플링
 * 처리 방법
   - 제거
   - 변환 (로그 등..), 구간화
   - 대체
   - 분할 분석

````python
import pandas as pd
gas1 = pd.read_csv('./data/gas_first_2019.csv', encoding='euc-kr')
gas2 = pd.read_csv('./data/gas_second_2019.csv', encoding='euc-kr')
gas = pd.concat([gas1, gas2], ignore_index=True)
````

````python
# 2sigma, 3sigma를 이용한 이상치 확인 방법
# 평균 +- 2std or 3td 범위 밖의 값은 outlier
mean = gas['휘발유'].mean()
std = gas['휘발유'].std()

# 2sigma
display(gas[(gas['휘발유']> mean+2*std) | (gas['휘발유']< mean-2*std) ])

# 3sigma
display(gas[(gas['휘발유']> mean+3*std) | (gas['휘발유']< mean-3*std) ])
````

````python

import seaborn as sns
tips = sns.load_dataset('tips')

plt.scatter(tips['tip'], tips['total_bill'])
````

````python
from scipy.stats import t
from statsmodels.formula.api import ols
````

````python
def hat_matrix(X):
    X = np.array(X)
    XtX = np.matmul(X.transpose(),X)
    XtX_inv = np.linalg.inv(XtX)
    return np.matmul(np.matmul(X, XtX_inv),X.transpose())
 
X = tips[['tip','total_bill']]
X['intercept'] =1
 
n = len(tips)
p = len(X.columns)
hat_mat = hat_matrix(X)
h = np.array([hat_mat[i][i] for i in range(n)])
 
mean_h = p/n
 
outlier_idx = np.where(h>2*mean_h)[0]
outlier = [h[i] for i in outlier_idx]

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot()
idx = range(1,n+1)
plt.scatter(idx, h)
plt.scatter(outlier_idx+1,outlier,s=200,facecolor='none',edgecolor='red')
for i in range(len(outlier)):
    ax.annotate(str(outlier_idx[i]+1),(outlier_idx[i]+1.5,outlier[i]),fontsize=13)
plt.axhline(2*mean_h,linestyle='--',color='red')
plt.xticks(idx)
plt.show()
````

````python
mean_x = np.mean(tips['tip'])
mean_y = np.mean(tips['total_bill'])
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot()
plt.xlabel('tip',fontdict={'size': 20})
plt.ylabel('total_bill',fontdict={'size': 20})
plt.scatter(tips['tip'], tips['total_bill'])
plt.scatter(mean_x,mean_y,alpha=0.3,color='green',s=200, label='Mean')
 
outlier_x = [tips['tip'][i] for i in outlier_idx]
outlier_y = [tips['total_bill'][i] for i in outlier_idx]
 
plt.scatter(outlier_x, outlier_y, s=200, facecolor='none', color='red', label='Outlier')
for i in range(len(outlier)):
    ax.annotate(str(outlier_idx[i]+1),(outlier_x[i]+0.5,outlier_y[i]),fontsize=13)
plt.legend()
plt.show()
````

#### 3-03-01 Boxplot

````python
# 값의 분포를 boxplot으로 확인함

sns.set_style("whitegrid")

plt.figure(figsize=(8,6))
sns.boxplot(x=tips["total_bill"])
plt.show()
````

### 3-04 heatmap
- 두 개의 카테고리 값에 의한 실수의 변화를 보고자 할 경우에 사용
- 표출하고자 하는 데이터를 먼저 데이터 테이블 형태로 만들어야 함

````python
# flights 데이터 세트 불러오기
import seaborn as sns
flights = sns.load_dataset("flights")
flights.head(5)
````

````python
# 해당 데이터를 heatmap으로 만들기 위해 'pivot' 함수를 사용하여 데이터 테이블 생성
flights = flights.pivot("month", "year", "passengers")
flights
````

````python
# heapmap 함수를 사용하여 데이터 테이블 시각화
plt.figure(figsize=(10,5))
sns.heatmap(flights)
plt.show()
````

````python
# annot 옵션을 통해 값 출력 가능
# fmt 옵션을 통해 데이터의 형식을 지정(d:정수, f:실수)
plt.figure(figsize=(10,8))
fig = sns.heatmap(flights, annot=True, fmt="d")
plt.show()
````

### 3-05 pairplot 
- 데이터 분포 및 변수간 관계를 분석

````python
# iris(붓꽃) 데이터를 불러옴
iris = sns.load_dataset("iris")
iris
````

````python
# pairplot을 이용하여 변수(컬럼)간 관계와 분포 확인 가능
sns.set(style="ticks")
sns.pairplot(iris)
plt.show()
````

````python
# hue 옵션을 이용하여 카테고리별 변수의 관계와 분포 확인 가능
sns.pairplot(iris, hue='species')
plt.show()
````

````python
# vars 옵션을 이용하여 특정 변수만 지정 가능
sns.pairplot(iris, vars=['sepal_width', 'sepal_length'])
plt.show()
````

````python
# vars 옵션을 이용하여 특정 변수만 지정 가능
sns.pairplot(iris, vars=['sepal_width', 'sepal_length'], hue='species')
plt.show()
````

````python
# x_vars, y_vars 옵션을 사용하여 해당 축의 특정 변수만 지정 가능
sns.pairplot(iris, x_vars=['sepal_width', 'sepal_length'],
            y_vars=['petal_width', 'petal_length'])
plt.show()
````

````python
# x_vars, y_vars 옵션을 사용하여 해당 축의 특정 변수만 지정 가능
sns.pairplot(iris, x_vars=['sepal_width', 'sepal_length'],
            y_vars=['petal_width', 'petal_length'], hue='species' )
plt.show()
````
