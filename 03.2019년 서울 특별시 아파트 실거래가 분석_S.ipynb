# 1. 법정동 코드 가공

```python
#법정동코드.xlsx을 읽어 locCode에 저장하세요
locCode = pd.read_excel('./dataset/법정동코드.xlsx')
locCode
```

```python
# 법정동 코드 컬럼에서 앞에서 5자리 추출합니다. 추출하여 '지역코드' 라는 새로운 컬럼에 입력하세요. 
# 힌트: 문자열로 변경하면 앞에서 5자리 추출이 가능. 법정동코드는 최종적으로 정수형이어야 함.
# 예) 서울특별시 종로구 > 11000

locCode['지역코드'] = locCode["법정동코드"].astype(str).str[0:5].astype('int64')
locCode.head()
```

```python
# 시도명이 서울특별시인 데이터만 추출 후, 시도명, 시군구명, 지역코드 컬럼만 추출하세요. 
# loc라는 변수에 저장하세요. 
loc = locCode[locCode['시도명']=='서울특별시'][['시도명','시군구명','지역코드']]
loc
```

# 2. 실거래가 데이터 가공

```python
# 실거래가 데이터 불러오기
# '2019년 서울 아파트 실거래가.csv' 데이터를 로드하세요. encoding="cp949"
apt = pd.read_csv('./dataset/2019년 서울 아파트 실거래가.csv', encoding="cp949")
apt.head()
```

```python
# 데이터 프레임의 컬럼을 확인합니다.
apt.columns
```

```python
# 데이터 프레임의 크기(차원)을 조회합니다
apt.shape
```

```python
# 불러온 데이터에 NA값이 있는지 조회합니다.
apt.isnull().sum()
```

```python
# 실거래가 데이터프레임 정보 조회
apt.info()
```

```python
# '거래금액' 컬럼의 콤마(,) 제거 후 데이터 타입을 정수형으로 변경하세요. 
apt['거래금액'] = apt['거래금액'].str.replace(',','').astype('int64')
apt.head()
```

```python
# 실거래가 데이터프레임 정보 조회 (거래금액 컬럼 변경 확인)
apt.info()
```

```python
# '전용면적'컬럼을 평으로 변환(나누기 3.306)하여 '전용면적(평)' 컬럼 생성 후 소숫점 둘쨰자리까지 표시합니다.
apt['전용면적(평)'] = round(apt['전용면적'] / 3.306, 2)
apt.head()
```

```python
# 전용면적을 범주화 하기 위해 전용면적(평) 컬럼을 정수형으로 바꾼 새로운 컬럼을 생성합니다.
apt['전용면적(평)2'] = apt['전용면적(평)'].astype('int64')
apt.head()
```

```python
# 아래의 코드를 실행하여 전용면적을 범주화 합니다.
# 힌트: cut 함수사용.   범위는 0 ~  ('전용면적(평)2'의 최대값/5 + 2 ) * 5 까지 5간격으로 범위를 설정하세요. 
daterange = np.arange(0,(int(max(apt['전용면적(평)2']) / 5) + 2) * 5, 5)
apt['전용면적(평)3'] = pd.cut(apt['전용면적(평)2'], daterange)
apt.head()
```

```python
# 범주화된 '전용면적(평)3'의 고유값을 확인합니다.
apt['전용면적(평)3'].unique()
```

```python
# loc 인덱서를 이용하여 거래월이 1~6인 경우 상반기, 7~12인 경우 하반기의 값을 가지는 컬럼을 생성합니다.('반기' 새컬럼생성)
apt.loc[apt['거래월'] <= 6, '반기'] ="상반기"
apt.loc[apt['거래월'] > 6, '반기'] ="하반기"
apt
```

```python
# loc 인덱서를 이용하여 거래월이 1~3월인 경우 1분기, 4~6월인경우 2분기,
# 7~9월인경우 3분기, 10~12월인 경우 4분기의 값을 가지는 컬럼을 생성합니다. ('분기' 새컬럼생성)
apt.loc[apt['거래월'] <= 3, '분기'] = "1분기"
apt.loc[(apt['거래월'] > 3) & (apt['거래월'] <= 6) , '분기'] ="2분기"
apt.loc[(apt['거래월'] > 6) & (apt['거래월'] <= 9) , '분기'] ="3분기"
apt.loc[(apt['거래월'] > 9) & (apt['거래월'] <= 12) , '분기'] ="4분기"
apt
```

```python
# 전용면적(평)2 컬럼을 이용하여 '평당거래금액' 컬럼을 생성합니다.(새컬럼생성)
# 평당거래금액 = 거래금액/전용면적(평)2

apt['평당거래금액'] = apt['거래금액'] / apt['전용면적(평)2']
apt.head()
```

# 3. 데이터 프레임 결합

```python
# '지역코드'를 기준으로 '법정동코드' 데이터프레임과 '아파트실거래가' 데이터프레임을 결합합니다.

df = pd.merge(apt, loc, on="지역코드")
df
```

```python
# 위에서 구한 df데이터프레임의 '지역코드' 컬럼을 삭제합니다.

df.drop(['지역코드'], axis=1, inplace=True)
df.head()
```

```python
# 위에서 구한 평균 거래 금액이 높은 상위 30개의 아파트 단지에 해당하는 거래 데이터를 이용하여 boxplot을 그려보세요.

apt_nm = df.groupby('아파트이름')['거래금액'].agg(['mean']).sort_values('mean', ascending=False).head(30).index
temp = df[df['아파트이름'].isin(apt_nm)]

plt.figure(figsize=(12,8))
sns.boxplot(x="아파트이름", y="거래금액", data=temp, palette="Set1")
plt.xticks(rotation=90)
plt.show()
```

```python
# 아파트이름별 평균 거래량(거래건수)이 높은 상위 30개의 아파트 단지를 추출하세요.
df.groupby('아파트이름')['거래금액'].agg(['count']).sort_values('count', ascending=False).head(30)
```

```python
# 위에서 구한 평균 거래량 높은 상위 30개의 아파트 단지에 해당하는 거래 데이터를 이용하여 boxplot을 그려보세요.

apt_nm = df.groupby('아파트이름')['거래금액'].agg(['count']).sort_values('count', ascending=False).head(30).index
temp = df[df['아파트이름'].isin(apt_nm)]

plt.figure(figsize=(12,8))
sns.boxplot(x="아파트이름", y="거래금액", data=temp, palette="Set1")
plt.xticks(rotation=90)
plt.show()
```
