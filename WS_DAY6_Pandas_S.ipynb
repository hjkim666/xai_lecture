```python
import pandas as pd
import numpy as np
```

 * data.csv 파일을 읽어서 df에 저장하세요

```python
df = pd.read_csv('ws_data.csv')
df
```

 - columns를 '날짜', '일', '수면', '밥', '수면', '영화', '게임'으로 변경


```python
df.columns = ['날짜', '일','수면','밥','휴식','영화','게임']
df
```

 * columns를 오름차순으로 정렬


```python
df = df.sort_index(axis=1)
df
```

 * df의 정보 조회

```python
df.info()
```

 - 게임 열의 값을 정수(int64)로 변경

```python
# 1차 변경 시도
# int64로 변경시 아래와 같이 오류가 나는 것이 정상입니다
df['게임'].astype('int64')
```

```python
# '게임'열의  '\t' 탭을 제거하여, '게임'열을 변경하세요.
# strip() 함수 이용  
df['게임'] = df['게임'].str.strip()
df
```

```python
# 2차 변경 시도 
# int64로 변경시 아래와 같이 오류가 나는 것이 정상입니다

df['게임'].astype('int64')
```

```python
# ',' 쉼표를 제거 하겠습니다.
# replace 함수를 이용하여 '게임' 열을 변경하세요. 
# str.replace(old, new)
df['게임'] = df['게임'].str.replace(',', '')
df
```

```python
# 3차 변경 시도
# int64로 변경시 아래와 같이 오류가 나는 것이 정상입니다

df['게임'].astype('int64')
```

```python
# 빈 문자열을 '0'(문자열)으로 치환해봅시다. 
# '게임' 열의 값이 공백('')인 값을 찾아 문자열'0'으로 변경하세요.  
df.loc[df['게임']=='', '게임'] = '0'
df
```

```python
# 4차 변경 시도
df['게임'].astype('int64')
# int64로 변경시 아래와 같이 오류가 나는 것이 정상입니다
```

```python
# float -> int 순으로 변경 해봅시다. 
# 먼저 '게임' 컬럼을 float64로 변경하고, 다시 int64로 변경하세요. 
df['게임'] = df['게임'].astype('float64').astype('int64')
df
```

```python
# 정보를 출력하여 변경 확인
df.info()
```

 - 수면 시간과 일한 시간의 비율을 수면/일로 계산하여 '수면비율'에 저장
