```python
import pandas as pd
 
encoding = 'latin1'
#engine = 'python'

upath ='movielens/users.dat'
rpath ='movielens/ratings.dat'
mpath ='movielens/movies.dat'

unames=['user_id', 'gender','age','occpuation','zip']
rnames=['user_id','movie_id','rating','timestamp']
mnames=['movie_id', 'title','genres']

users=pd.read_csv(upath, sep='::',header=None, names=unames,  engine='python'
                  , encoding=encoding)
ratings=pd.read_csv(rpath, sep='::',header=None, names=rnames,  engine='python'
                    , encoding=encoding)
movies=pd.read_csv(mpath, sep='::',header=None, names=mnames,  engine='python'
                   , encoding=encoding)

users[:5]
```

```python
movies[:5]
```

```python
ratings
```

```python
# 테이블 병합
data = pd.merge(pd.merge(ratings, users), movies)
data
```

```python
data.iloc[0]
```

```python
# 성별에 따른 각 영화의 평균 평점 -> pivot_table 을 이용 
mean_ratings = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
mean_ratings
```

```python
mean_ratings[:5]
```

```python
# 250건이상 평점정보가 있는 영화만 추리기 
ratings_by_title = data.groupby('title').size()
ratings_by_title
```

```python
ratings_by_title[:10]
```

```python
active_titles = ratings_by_title[ratings_by_title >= 250]
active_titles
```

```python
# 250건이상 평점정보가 있는 영화에 대한 색인 작성 -> mean_ratings 에서 추출 
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings
```

```python
# 평점 차이 구하기 
# 남녀간의 호불호가 갈리는 영화를 찾아보자 

mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
mean_ratings
```

```python
# 남녀간의 호불호가 갈리는 영화를 정렬하자 
# 정렬된 결과를 정렬 순으로 15건 추출 

sorted_by_diff = mean_ratings.sort_values(by='diff')
sorted_by_diff[:15]
```

```python
# 남성들이 더 선호하는 순 -> 역순으로 정렬한다음 상위 15개 행으로 잘라냄 
sorted_by_diff[::-1][:15]
```

##### 성별에 관계없이 호불호가 극명하게 나뉘는 영화 찾기 
- 분산 or 표준편차가 큰 데이터를 찾아 호불호가 갈리는 데이터를 찾아봅시다. 


```python
# 성별에 관계없이 호불호가 극명하게 나뉘는 영화 -> 분산 or 표준편차 

# 평점의 표준편차 
rating_std_by_title = data.groupby('title')['rating'].std()

#active_titles만 선택 
rating_std_by_title = rating_std_by_title.loc[active_titles]

# 내림 차순으로 정렬 
rating_std_by_title.sort_values(ascending=False)[:10]
```
