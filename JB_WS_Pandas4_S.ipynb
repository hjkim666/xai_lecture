```python
# 기온의 평균, 표준편차, 최고, 최저 값을 각각 출력하세요.
# 소수점 셋째자리까지 출력하세요.
print(f'평균 기온: {df["temperature"].mean():.3f}')
print(f'표준편차: {df["temperature"].std():.3f}')
print('최고 기온:', df['temperature'].max())
print('최저 기온:', df['temperature'].min())
```

```python
# 기온의 평균, 최고, 최저 값을 시리즈로 출력하세요. 
# 소수점 셋째자리까지 출력하세요.
df['temperature'].agg(['mean', 'max','min']).round(3)
```

```python
# 풍속의 평균, 모분산, 표본분산을 출력하세요.
def var0(x):
    return x.var(ddof=0)

df['windspeed'].agg(평균='mean', 모분산=var0, 표본분산='var')
```

```python
# 풍속의 평균, 모분산, 표본분산을 출력하세요.
def var0(x):
    return x.var(ddof=0)

df['windspeed'].agg(평균='mean', 모분산=lambda x: x.var(ddof=0), 표본분산='var')
```

```python
# 습도의 중앙값, 최소, 최대, 범위를 출력하세요.
def x_range(x):
    return x.max() - x.min()

df['humidity'].agg(중앙값=('median'), 최댓값=('max'), 최솟값=('min'), 범위=(x_range))
```

```python
# 가시거리의 평균, 표준편차(표본), 중앙값, IQR, 최소, 최댓값을 출력하세요.
# 소수점 넷째자리까지 출력하세요.
def IQR(x):
    return x.quantile(.75) - x.quantile(.25)

df['visibility'].agg(평균=('mean'),
                     표준편차=('std'),
                     중앙값=('median'),
                     IQR=(IQR), 최솟값=('min'), 최댓값='max').round(4)
```

```python
# IQR 방식을 이용하여 가시거리에 이상치(outlier)가 있는지 확인하세요.
Q1 = df['visibility'].quantile(.25)
Q3 = df['visibility'].quantile(.75)
IQR = Q3-Q1
tukey_max = Q3 + 1.5*IQR
tukey_min = Q1 - 1.5*IQR

df[(df['visibility'] > tukey_max) | (df['visibility'] < tukey_min)]
```

```python
# IQR 방식을 이용하여 풍속에 이상치(outlier)가 있는지 확인하세요.
Q1 = df['windspeed'].quantile(.25)
Q3 = df['windspeed'].quantile(.75)
IQR = Q3-Q1
tukey_max = Q3 + 1.5*IQR
tukey_min = Q1 - 1.5*IQR

df[(df['windspeed'] > tukey_max) | (df['windspeed'] < tukey_min)]
```

```python
# 오존, 미세먼지, 초미세먼지의 평균, 표준편차(표본), 최소, 최댓값을 출력하세요.
df[['ozone', 'pm10', 'pm2.5']].agg(['mean', 'std', 'min', 'max'])
```

```python
# 미세먼지나 초미세먼지가 평균치를 초과했던 데이터를 추출하여 출력하고,
# 전체 데이터 중 몇 퍼센트에 해당하는지 출력하세요.
pm10_mean = df['pm10'].mean()
pm25_mean = df['pm2.5'].mean()

df1025초과 = df[(df['pm10']>pm10_mean)|(df['pm2.5']>pm25_mean)]
display(df1025초과)
print(f'미세먼지나 초미세먼지가 평균치를 초과했던 경우는 전체 {len(df)}건 중 {len(df1025초과)}건으로 약 {len(df1025초과)/len(df)*100:.2f}%입니다.')
```

```python
# 미세먼지와 초미세먼지가 모두 평균치를 초과했던 날의 데이터를 추출하세요.
pm10_mean = df['pm10'].mean()
pm25_mean = df['pm2.5'].mean()

df1025초과 = df[(df['pm10']>pm10_mean)&(df['pm2.5']>pm25_mean)]
display(df1025초과)
print(f'미세먼지와 초미세먼지가 모두 평균치를 초과했던 경우는 전체 {len(df)}건 중 {len(df1025초과)}건으로 약 {len(df1025초과)/len(df)*100:.2f}%입니다.')
```

```python
# 강수 여부 변수의 결측치를 습도가 80.35% 이상이면 'yes'
# 80.35% 미만이면 'no'로 대체하세요.
# 습도가 결측치인 경우는 그대로 결측치로 유지하세요.
df_prena = df.loc[df['precipitation'].isna()]
df_prena.loc[df['humidity']>=80.35, 'precipitation'] = 'yes'
df_prena.loc[df['humidity']<80.35, 'precipitation'] = 'no'
df.loc[df_prena.index, 'precipitation'] = df_prena['precipitation']
df
```
