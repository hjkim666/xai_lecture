#### 1-01-03 스타일 지정
- plt.plot(x, y, style)

  image.png

````python
# plt.plot(x, y, style) 함수를 사용하여 다음과 같은 그래프를 그려본다
# x, y 축을 갖고, 빨간색 원('ro')으로 표시되는 그래프
# x 축 0, 1,2,3,...9
# y 축 x ** 2
x = np.arange(10)
y = x ** 2
plt.plot(x,y, 'y^')
````

#### 1-01-04 여러 그래프 겹쳐 그리기
- plt.plot(x, y, style, x2, y2, style2, ...)



````python
# 여러 그래프 겹쳐 그리기
# plt.plot(x, y, style, x2, y2, style2, ...)
x = np.arange(10)
plt.plot(x, x*2, 'b-', x, x**2, 'r--', x, x**3, 'go')
````

````python
# 위의 그래프를 plt.plot(x, y, style)을 여러번 사용하여 겹쳐 그리기로 작성해 보도록 한다
x = np.arange(10)
plt.plot(x, x*2, 'b--', label='A')
plt.plot(x, x**2, 'r--', label='B')
plt.plot(x, x**3, 'go', label='C')
plt.legend()
````

#### 1-01-05 Axis Label, Title
- plt.xlabel(str) : x 축에 대한 레이블 표시
- plt.ylabel(str) : y 축에 대한 레이블 표시
- plt.title(str) : plot에 대한 제목

````python
# 다음 출력과 같이 레이블과 제목을 표시하도록 한다.
x = np.arange(10)
plt.plot(x, x*2, 'b-', x, x**2, 'r--', x, x**3, 'go')
plt.xlabel('money')
plt.ylabel('time')
plt.title('time for money')
````

#### 1-01-06 Limitation
- 전체 데이터 중에 특정 영역만 표시
- plt.xlim(min, max)
- plt.ylim(min, max)

````python
# 전체 데이터 중 특정 영역만 표시해 본다
x = np.arange(10)  # 0~9
y = x ** 2         # 0~81
plt.plot(x, y)

# x 축에 대해서 2~5, y축에 대해서 5~20 영역 표시
plt.xlim(2, 5)
plt.ylim(5, 20)
````

#### 1-01-07 Legend(범례)
- plt.plot(x, y, style, label='text')
- plt.legend(loc='text')
   - loc : upper left, upper center, upper right, center left, center, center right, lower left, lower right, best, right 등이 있음

````python
# Legend(범례)
x = np.arange(10)
y = x ** 2
y2 = x*5 + 2
# 다음과 같은 결과로 범례가 표시되도록 plt.plot의 label, plt.legend의 loc을 지정해 보도록 한다
plt.plot(x, y, 'b', label='first')
plt.plot(x, y2, 'r', label='second')
plt.legend(loc='upper right')
````

#### 1-01-08 Annotation
- plt.annotate(text, xy, xytext, arrowprops)
    - text : 표시할 문구
    - xy : 화살표를 표시할 좌표
    - xytext : 문구를 표시할 좌표
    - arrowprops : 화살표 속성
- plt.text(x, y, text)
    - x, y : 문구 표시할 위치 좌표
    - text : 표시할 문구

````python
# annotation & text

x = np.arange(10)
y = x ** 2
plt.plot(x, y)
plt.annotate('here', 
             xy = (4, 16),
             xytext=(5, 20),
             arrowprops={'color' : 'green'})
plt.text(7, 70, '$y=x^2$')
````

````python
# other example

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
plt.plot(t, s, lw=2)
plt.annotate('local max', xy=(2,1), xytext=(2.5,1.2),
            arrowprops=dict(facecolor='black', shrink=0.1))

plt.ylim(-2,2)
plt.show()
````

### 1-02 Subplot
- 여러 plot을 나누어 표시
- plt.subplot(row, col, idx)
    - row, col : shape(행, 열)
    - idx : 일련번호 1부터 시작

````python
# subplot

x = np.arange(0, 10, 0.1)

plt.subplot(2, 2, 1)
plt.plot(x, x**2)

plt.subplot(2, 2, 2)
plt.plot(x, x*5)

plt.subplot(223)
plt.plot(x, np.sin(x))

plt.subplot(224)
plt.plot(x, np.cos(x))
````

### 1-06.히스토그램 (Histogram)
- 히스토그램을 그리기 위한 hist 명령은 bins 인수로 데이터를 집계할 구간 정보를 받음
- 데이터 집계 결과를 반환
- n, bins, patches = plt.hist(x, bins, range)
    - x : 데이터
    - bins : 계급
    - range(대상 범위)
    - n : 도수
    - patches : 계급별 히스토그램 좌표와 높이

````python
x = 100 + 15 *np.random.randn(10000)
# n, bins, patches = plt.hist(x, 50)
plt.hist(x, 50)
plt.show()
````

````python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)
#print(n, bins, patches, sep='\n')
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.text(60, 0.025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
````

### 3-01 Line Plot


````python
# 라인플롯(Line plot) 그리기
x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = 2 * np.sin(x+0.5)
y3 = 3 * np.sin(x+1.0)
y4 = 4 * np.sin(x+1.5)

plt.figure(figsize=(10,6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
````

### 3-03 Tips Data를 활용한 시각화 실습

````python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline

sns.set_style("whitegrid")
tips = sns.load_dataset("tips")
tips.head(5)
#type(tips)
````

#### 3-03-01 Boxplot

````python
# 값의 분포를 boxplot으로 확인함

sns.set_style("whitegrid")

plt.figure(figsize=(8,6))
sns.boxplot(x=tips["total_bill"])
plt.show()
````

````python
# 특정 카테고리에 대한 값의 분포는 X,Y를 지정하여 확인 가능
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
````

````python
# hue 옵션을 통해 카테고리 별 데이터 시각화 가능
# palette 옵션을 통해 색상 지정 가능
plt.figure(figsize=(8,6))

sns.boxplot(x="day", y="tip", hue="smoker",data=tips, palette="Set2")
plt.show()
````

#### 3-03-02 swarm plot
- boxplot 보다 원본 데이터의 분포를 더 자세하게 확인할 때 사용

````python
plt.figure(figsize=(8,6))
sns.swarmplot(x="day", y="total_bill", data=tips, color=".1")
plt.show()
````

````python
# boxplot + swarmplot
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
plt.show()
````

````python
# Implot를 이용하여 회귀 그래프를 생성할 수 있음
sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", data=tips)
plt.show()
````

````python
# hue 옵션을 통해 카테고리 별 적용 가능
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips)
plt.show()
````

````python
# palette 옵션을 통해 색상 적용 가능
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, palette="Set1")
plt.show()
````

### 3-04 heatmap
- 두 개의 카테고리 값에 의한 실수의 변화를 보고자 할 경우에 사용
- 표출하고자 하는 데이터를 먼저 데이터 테이블 형태로 만들어야 함

````python
# flights 데이터 세트 불러오기
flights = sns.load_dataset("flights")
flights.head(5)
````

````python
# 해당 데이터를 heatmap으로 만들기 위해 'pivot' 함수를 사용하여 데이터 테이블 생성
flights = flights.pivot("month", "year", "passengers")
flights
````

````python
# heapmap 함수를 사용하여 데이터 테이블 시각화
plt.figure(figsize=(10,5))
sns.heatmap(flights)
plt.show()
````

````python
# annot 옵션을 통해 값 출력 가능
# fmt 옵션을 통해 데이터의 형식을 지정(d:정수, f:실수)
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt="d")
plt.show()
````

### 3-05 pairplot 
- 데이터 분포 및 변수간 관계를 분석

````python
# iris(붓꽃) 데이터를 불러옴
iris = sns.load_dataset("iris")
iris
````

````python
# pairplot을 이용하여 변수(컬럼)간 관계와 분포 확인 가능
#sns.set(style="ticks")
sns.pairplot(iris)
plt.show()
````

````python
# hue 옵션을 이용하여 카테고리별 변수의 관계와 분포 확인 가능
sns.pairplot(iris, hue='species')
plt.show()
````

````python
# vars 옵션을 이용하여 특정 변수만 지정 가능
sns.pairplot(iris, vars=['sepal_width', 'sepal_length'], hue='species')
plt.show()
````

````python
# x_vars, y_vars 옵션을 사용하여 해당 축의 특정 변수만 지정 가능
sns.pairplot(iris, x_vars=['sepal_width', 'sepal_length'],
            y_vars=['petal_width', 'petal_length'], hue='species' )
plt.show()
````
